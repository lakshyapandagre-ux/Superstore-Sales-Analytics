```python
pareto_df = (
    df.groupby(
        'Customer Name'
    )['Sales']
    .sum()
    .sort_values(
        ascending=False
    )
    .reset_index()
)

pareto_df.head()
```

```python
pareto_df['Cumulative_Sales'] = (
    pareto_df['Sales']
    .cumsum()
)

pareto_df['Cumulative_Percentage'] = (
    pareto_df['Cumulative_Sales']
    /
    pareto_df['Sales'].sum()
) * 100
```

```python
fig = px.line(
    pareto_df,
    x=pareto_df.index,
    y='Cumulative_Percentage',
    title='Customer Revenue Pareto Analysis'
)

fig.show()
```
